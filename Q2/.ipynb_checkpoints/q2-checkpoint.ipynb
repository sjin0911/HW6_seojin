```python
import csv
import matplotlib.pyplot as plt
import random as rand

def main():
    f=[]
    f.append(open('100.csv','w'))
    f.append(open('1000.csv','w'))
    f.append(open('10000.csv','w'))
    f.append(open('100000.csv','w'))

    num=[100, 1000, 10000, 100000]

    for i in range(0,len(f)):
        n=10**(i+2)
        f[i].write("n={0:d}\n".format(n))
        for j in range(0,n):
            data="{0:d},{1:d}\n".format(j+1,rand.randrange(1,7))
            f[i].write(data)
        f[i].close()
    
    file=[]
    result=[[],[],[],[]]
    file.append(open('100.csv','r',encoding='cp949'))
    file.append(open('1000.csv','r',encoding='cp949'))
    file.append(open('10000.csv','r',encoding='cp949'))
    file.append(open('100000.csv','r',encoding='cp949'))
        
    for i in range(0,len(file)):
        data=csv.reader(file[i])
        next(data)
        for row in data:
            result[i].append(int(row[1]))
    
    for i in range(0,len(result)):
        plt.title("n={0:d}".format(10**(i+2)))
        plt.hist(result[i],bins=11)
        plt.show()

if __name__ == '__main__':
    main()
```
