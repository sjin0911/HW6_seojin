```python
import csv
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

def main():
    
    in_out={}
    in_={}
    out_={}
    value=[[],[],[]]
    rc('font', family='AppleGothic')

    f=open('seoulMetro.csv','r',encoding='utf-8')
    data=csv.reader(f)
    next(data)
    next(data)
    
    for row in data:
        if in_out.get(row[3])==None:
            in_out[row[3]]=int(row[10])+int(row[11])+int(row[12])+int(row[13])
            in_[row[3]]=int(row[10])+int(row[12])
            out_[row[3]]=int(row[11])+int(row[13])
        else:
            in_out[row[3]]=in_out.get(row[3])+int(row[10])+int(row[11])+int(row[12])+int(row[13])
            in_[row[3]]=in_.get(row[3])+int(row[10])+int(row[12])
            out_[row[3]]=out_.get(row[3])+int(row[11])+int(row[13])
            
    in_out_sort=sorted(in_out.items(),key=lambda x:x[1],reverse=True)
    in_sort=sorted(in_.items(),key=lambda x:x[1],reverse=True)
    out_sort=sorted(out_.items(),key=lambda x:x[1], reverse=True)

    in_out_sort=np.array(in_out_sort)
    in_sort=np.array(in_sort)
    out_sort=np.array(out_sort)
    
    for i in range(0,30):
        value[0].append(int(in_out_sort[i][1]))
        value[1].append(int(in_sort[i][1]))
        value[2].append(int(out_sort[i][1]))
            
    plt.figure(figsize=(50, 30))
    plt.bar(in_out_sort[0:30,0],value[0])
    plt.title('Most Used station (IN and OUT)')
    plt.xlabel('Name of station')
    plt.ylabel('Number of users')
    
    plt.figure(figsize=(50, 30))
    plt.bar(in_sort[0:30,0],value[1])
    plt.title('Most Used Station (only IN)')
    plt.xlabel('Name of station')
    plt.ylabel('Number of users')
    
    plt.figure(figsize=(50, 30))
    plt.bar(out_sort[0:30,0],value[2])
    plt.title('Most Used Station (only OUT)')
    plt.xlabel('Name of station')
    plt.ylabel('Number of users')
    
    plt.show()

if __name__=='__main__':
    main()
```
